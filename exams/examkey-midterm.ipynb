Use a bar plot to show the expected probability distribution of 0-20 lever presses in a minute. *Hint: Is the expected distribution continuous or discrete?
```python
import scipy.stats as st
import matplotlib.pyplot as plt

numLeverPresses = np.arange(21)
leverPressesPerMinuteDist = st.poisson(5)

plt.bar(numLeverPresses, leverPressesPerMinuteDist.pmf(numLeverPresses))
plt.xlabel('# Lever Presses per Minute')
plt.ylabel('Probability');
```

```python
all_ms = np.hstack([control_ms, toxin_ms])
perm_diffs_ms = np.zeros((10000))
for i in range(10000):
    np.random.shuffle(all_ms)
    perm_a = all_ms[:len(control_ms)]
    perm_b = all_ms[-len(toxin_ms):]
    perm_diffs_ms[i] = perm_b.mean() - perm_a.mean()

lb, ub = np.quantile(perm_diffs_ms, [0.025, 0.975])

print(f"""
Assuming the toxin has no effect on mean time to opening following pulse onsset, 
we should measure differences in mean time to opening between control and toxin samples 
that fall between {lb:.2f} and {ub:.2f} ms with 95% confidence.
""")
```
